## Task 2.1:
Introducing our own feature. <br>
Mean/average of all scores:
(math_score + reading_score + writing_score)/3 = STUDENT_AVG_SCORE

- LOW
- BELOW MEDIUM
- MEDIUM
- ABOVE MEDIUM
- HIGH



0 --> 40 --> LOW <br>
41 --> 50 --> BELOW MEDIUM <br>
51 -->  75 --> MEDIUM <br>
76 --> 85 --> ABOVE MEDIUM<br>
86 --> 100 --> HIGH<br>

We'll create bins for each of these.

```python
sp['AVG'] = ((sp['math score'] + sp['writing score'] + sp['reading score']) / 3).round(decimals=2)
```

```python
sp.head()
```

```python
def categorize_score(avg_score):
    if avg_score <= 40:
        return 'LOW'
    elif 41 <= avg_score <= 50:
        return 'BELOW MEDIUM'
    elif 51 <= avg_score <= 75:
        return 'MEDIUM'
    elif 76 <= avg_score <= 85:
        return 'ABOVE MEDIUM'
    elif 86 <= avg_score <= 100:
        return 'HIGH'
    else:
        return 'Out of range'
```

```python
sp['AVG Category'] = sp['AVG'].apply(categorize_score)
```

```python
sp.head()
```

```python
# Defining bins for categorizing average scores
bins = [0, 40, 50, 75, 85, 100]
labels = ['LOW', 'BELOW MEDIUM', 'MEDIUM', 'ABOVE MEDIUM', 'HIGH']

# new column 'AVG Category' based on bins
sp['AVG Category'] = pd.cut(sp['AVG'], bins=bins, labels=labels, include_lowest=True)

# Count the number of students in each category
category_counts = sp['AVG Category'].value_counts()

plt.bar(category_counts.index, category_counts.values, color='skyblue')
plt.xlabel('Average Score Category')
plt.ylabel('Number of Students')
plt.title('Distribution of Average Score Categories')
plt.show()
```

Introducing one-hot-encoding for our 'race/ethnicity' category.

```python
sp["race/ethnicity"].unique()
```

```python
sp_encoded = pd.get_dummies(sp, columns=['parental level of education', 'lunch', 'race/ethnicity'])
sp_encoded.head()
```

Above we used the pandas function **.get_dummies** to perform one-hot-encoding on the categorical variables: <br>
- parental level of education
- lunch
- race/ethnicity

We see that every row will get every possible variable for each of the categories mentioned above. <br>
For example the lunch has two possible values:
- free/reduced
- standard

One-hot-encoding does binary encoding, which assigns True/False depending on the actual value. <br>
This will be more efficient and smart for our machine learning model in the future, since assigning 2 numbers for the differents possibilities wouldn't have much meaning, and also does that mean that free/reduced is better or vice versa? Also this is actually a must since the machine learning model wouldn't be able to understand strings anyway. <br>

<br>
Without one-hot encoding, categorical variables are often encoded as integers, which could imply an ordinal relationship between categories where none exists. One-hot encoding prevents such misinterpretations by representing each category independently. <br><br>
By accurately representing categorical variables, one-hot encoding can lead to better model performance. It allows algorithms to identify patterns and relationships within categorical features more effectively, potentially leading to more accurate predictions.<br> <br>
One-hot encoding reduces bias that might be introduced by encoding categorical variables in a different manner. It ensures that each category receives equal representation in the model's learning process.
