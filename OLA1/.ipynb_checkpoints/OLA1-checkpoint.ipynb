## Data Visualization

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.catplot(data=sp)
```

The plot above shows a scatter plot, which shows the spread of scores depending on the different subjects

```python
sns.catplot(data=sp, kind="box")
```

This plot above displays 3 boxplots which show the min, max, 25th precentile, 50th precentile aka the median and 75th precentile, as well as outliers wihtin the different subjects.

```python
sns.catplot(data=sp, x="math score", y="gender", kind="box")
```

The plot above displays 2 boxplots showing math scores for each gender.

```python
sns.catplot(data=sp, x="math score", y="test preparation course", kind="box") #...
```

The plot above displays 2 boxplots showing math scores depending on if you had done a test preparation course prior to the acutal test.

```python
sns.catplot(data=sp, x="math score", y="parental level of education", kind="box") #...
```

This plot displays multiple boxplots showing the math scores based on the parental level of education.
We can assume that students with parents having higher levels of education perform marginally better.

```python
sns.catplot(data=sp, x="math score", y="parental level of education", hue="gender", kind="box")
```

This plot displays multiple boxplots showing the **math** scores based on the parental level of education now separated by genders. We can see that male students usually perform better than female students in math.

```python
sns.catplot(data=sp, x="reading score", y="parental level of education", hue="gender", kind="box") #..
```

This plot displays multiple boxplots showing the **reading** scores based on the parental level of education now separated by genders. We now can see/assume that the female students perform better than the male counterpart in reading.

```python
sns.catplot(data=sp, x="writing score", y="parental level of education", hue="gender", kind="box") #..
```

This plot displays multiple boxplots showing the **writing** scores based on the parental level of education now separated by genders. We now can see/assume that the female students perform better than the male counterpart in writing as well.

```python
sns.catplot(data=sp, x="math score", y="parental level of education", hue="race/ethnicity", kind="box") #..
```

Here we see that the race/ethnicity **GROUP E** typically perform better.

```python
import warnings
warnings.filterwarnings('ignore')
sns.catplot(data=sp, x="math score", y="parental level of education", hue="gender", kind="swarm")
```

This is another scatter plot that displays the student math scores based on their parental level of education showing their genders on the plot

we can see based on our visualizations that there is a correlation between genders, their parental level of education and eventually their differents scores. This does not lead to a causation per se. There might be other external factors playing as well.
