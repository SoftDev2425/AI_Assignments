# Data Exploration

```python
import pandas as pd
```

```python
# store the .csv in sp variable
sp = pd.read_csv("./OLA1/data/StudentsPerformance.csv")
```

```python
sp.head() # This display the first 5 rows
```

```python
sp["parental level of education"].unique()
```

```python
sp.tail() # This displays the last 5 rows
```

```python
sp.describe() # This displays info about our numerical data in the dataset
```

In the above table we see statistics for the columns **math score**, **reading score** and **writing score**.

We can read from the table that the max scores for all columns is 100 (great success).

We can also read the the mean, which is the average for each of the columns.

We can also read the std (standard deviation) for each column, which tells us the spread of values in relation to the mean (avg), which in this case is around 15 for each of the columns.

We've also got the minimum and the maximum values in each of the columns.

Also the table tells us that 25% of the total math scores were 57 or less, reading score 59 or less and writing score 57.75 or less.

Also the table tells us that 50% aka the median of the total math scores were 66, reading score 70 and writing score 69 or less.

Also the table tells us that 75% of the total math scores were 77 or less, reading score 79 or less and writing score 79 or less.

## Data Visualization

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.catplot(data=sp)
```

```python
sns.catplot(data=sp, kind="box") #...
```

```python
sns.catplot(data=sp, x="math score", y="gender", kind="box") #...
```

```python
sns.catplot(data=sp, x="math score", y="test preparation course", kind="box") #...
```

```python
sns.catplot(data=sp, x="math score", y="parental level of education", kind="box") #...
```

```python
sns.catplot(data=sp, x="math score", y="parental level of education", hue="gender", kind="box")
```

```python
sns.catplot(data=sp, x="reading score", y="parental level of education", hue="gender", kind="box") #..
```

```python
sns.catplot(data=sp, x="writing score", y="parental level of education", hue="gender", kind="box") #..
```

```python
sns.catplot(data=sp, x="math score", y="parental level of education", hue="race/ethnicity", kind="box") #..
```

we can see based on out visualizations that there is a correlation between genders, their parental level of education and eventually their differents scores. This does not lead to a causation per se. There might be other external factors playing as well.
